```python
import random
import matplotlib.pyplot as plt
import numpy as np

def biased_toss(x = None):
    """A biased coin toss, the bias is set by the variable x which represents the 
    probability of a heads (1). The probability of a tails (0) is therefore 1-x
    """
    if x is None:
        x = 0.8
    value = random.random()
    if value < x:
        return 1
    return 0

def naive_unbiased_toss():
  """Returns 1 for heads, 0 for tails. 50/50 chance of either"""
  return (biased_toss() - biased_toss()) < 0

def unbiased_toss():
  """Returns 1 for heads, 0 for tails. 50/50 chance of either"""
  difference = biased_toss() - biased_toss()
  if difference < 0:
    return 0
  if difference > 0:
    return 1
  return unbiased_toss()
```

```python
total_sum = 0
values = []
for i in range(1, 10001):
    total_sum += biased_toss()
    values.append(total_sum)
```

```python
x_axis = [i + 1 for i in range(len(values))]
y_axis = [values[i] / (i + 1) for i in range(len(values))]
plt.plot(x_axis, y_axis)
plt.plot(x_axis, [0.8]*len(x_axis))
plt.xscale('log')
```

```python
x = np.arange(0, 1, 1/1000)
prob_func = lambda x: x**2 + (1 - x)**2
y = [prob_func(i) for i in x]
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0.1, 0.9, 1/1000)
expected_value_func = lambda x: 1 / (1 - (x**2 + (1 - x)**2))
y = [expected_value_func(i) for i in x]
plt.plot(x, y)
plt.show()
```
